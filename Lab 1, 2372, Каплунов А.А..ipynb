# Описание набора данных

## 1. Предметная область
**Медицина** — данные касаются различных параметров здоровья пациентов и факторов риска сердечно-сосудистых заболеваний.

## 2. Источник данных
**Источник**: [Kaggle](https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset) — датасет был скачан с платформы Kaggle, где он доступен в открытом доступе.

## 3. Характер данных
**Тип данных**: Реальные данные, собранные из медицинских записей пациентов.

## 4. Описание атрибутов датасета

| Имя столбца | Тип данных | Описание                            |
|-------------|------------|-------------------------------------|
| `age`       | Числовой   | Возраст пациента                   |
| `sex`       | Категориальный | Пол пациента (0 — женский, 1 — мужской) |

| `fbs`       | Категориальный | Сахар в крови натощак (>120 мг/дл) |
| `restecg`   | Категориальный | Результаты электрокардиографии   |
| `thalach`   | Числовой   | Максимальный пульс                 |

| `oldpeak`   | Числовой   | Депрессия ST сегмента              |
| `slope`     | Категориальный | Наклон ST сегмента               |
| `ca`        | Числовой   | Количество крупных сосудов         |
| `thal`      | Категориальный | Статус талассемии                |
| `target`    | Категориальный | Наличие сердечного заболевания (0 — нет, 1 — да) |


## 5. Описание задачи анализа
Основная задача — классификация пациентов по наличию или отсутствию сердечно-сосудистых заболеваний, чтобы выявить группы риска и потенциальные факторы, влияющие на здоровье сердца.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown
%matplotlib inline 
```

```python
df = pd.read_csv('heart.csv')

def analyze_attribute(df, column_name):
    mean_value = df[column_name].mean()
    std_value = df[column_name].std()
    missing_count = df[column_name].isnull().sum()
    # Если пропуски есть, заполним их медианой (здесь предположим, что пропусков нет):
    # df['column_name'].fillna(df['column_name'].median(), inplace=True)
    
    # Определение выбросов с помощью метода интерквартильного размаха
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_count = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)].shape[0]

    analysis_text = f"""
### Анализ атрибута: `{column_name}`

- **Среднее значение**: {mean_value:.2f}
- **Стандартное отклонение**: {std_value:.2f}
- **Пропущенные значения**: {missing_count}
- **Количество выбросов**: {outliers_count}
"""
    
    display(Markdown(analysis_text))
```

```python
for attribute in df.select_dtypes(include=[np.number]).columns:
    analyze_attribute(df, attribute)
```

```python
for column in df.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True)
    plt.title(f"Распределение значений для {column}")
    plt.xlabel(column)
    plt.show()
```

```python
correlation_matrix = df.corr()

print("Матрица корреляции:")
print(correlation_matrix)

threshold = 0.5  # Порог для определения высокой корреляции
high_corr_pairs = correlation_matrix[(correlation_matrix >= threshold) & (correlation_matrix < 1.0)]

print("\nВысококоррелированные атрибуты:")
for i in range(len(high_corr_pairs.columns)):
    for j in range(i):
        if abs(high_corr_pairs.iloc[i, j]) >= threshold:
            print(f"{high_corr_pairs.columns[i]} и {high_corr_pairs.columns[j]}: {high_corr_pairs.iloc[i, j]:.2f}")
```

```python
plt.figure(figsize=(12, 10))
pair_plot = sns.pairplot(df)
pair_plot.fig.suptitle('Матрица графиков рассеивания', y=1.02)  # Установка заголовка для всей фигуры
plt.show()
```

### Анализ корреляции между параметрами

1. **Матрица корреляции:**
   Матрица корреляции показывает, как каждый атрибут связан с другими атрибутами. Значения варьируются от -1 до 1:
   - Значение **1** указывает на полную положительную корреляцию.
   - Значение **-1** указывает на полную отрицательную корреляцию.
   - Значение **0** указывает на отсутствие корреляции.

2. **Высококоррелированные атрибуты:**
   - В результате анализа не было обнаружено высококоррелированных атрибутов (корреляция выше 0.5 или ниже -0.5).

3. **Не имеющие корреляции атрибуты:**
   - Многие атрибуты не имеют значительной корреляции друг с другом, включая такие пары как:
        - **age и sex**: корреляция -0.103
        - **trestbps и chol**: корреляция 0.127
        - **restecg и thalach**: корреляция 0.048
        - **и другие с корреляцией, близкой к нулю.**

### Графики рассеивания
- Матрица графиков рассеивания предоставляет визуальное представление зависимостей между числовыми атрибутами. На графиках можно увидеть:
  - **Положительные корреляции**: Когда точки графика образуют восходящую линию.
  - **Отрицательные корреляции**: Когда точки образуют нисходящую линию.
  - **Отсутствие корреляции**: Когда точки расположены случайным образом.

### Выводы
- Не было выявлено сильной корреляции между атрибутами, что может указывать на то, что большинство параметров в наборе данных являются независимыми друг от друга.
- На графиках рассеивания видно, что точки распределены достаточно случайно, что подтверждает отсутствие значительной линейной зависимости между переменными.
- Особое внимание стоит уделить атрибуту target, который имеет некоторую корреляцию с exang (-0.438) и oldpeak (-0.438), что может свидетельствовать о наличии некоторой связи с целевым признаком.
